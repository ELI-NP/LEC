```python
%matplotlib notebook
# use ``%matplotlib widget`` in Jupyter Lab
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import matplotlib.gridspec as gridspec
from matplotlib.colors import LogNorm
matplotlib.axes.Axes.annotate
matplotlib.pyplot.annotate
import pathlib
import os
import pwd

def get_username():
    return pwd.getpwuid(os.getuid())[0]
user = get_username()

run_dir  = pathlib.Path(rf"/home/{user}/LEC/examples/")

print(f"run_dir: {run_dir}")
```

```python
import matplotlib as mpl
import figformat 
fig_width,fig_height,params=figformat.figure_format(fig_width=3.4,fig_height=2)
mpl.rcParams.update(params)
```

```python
file1=1
file2=2
fig, ax1 = plt.subplots()

#Time evolution of electron energy
T1,x1,y1,px1,py1,E1,TT1,RAD1,Xi1 = np.loadtxt(rf"{run_dir}/Data1/orbt1q"
                                              +str(file1).zfill(3)+".dat",unpack=True,
                                              usecols=[0,1,2,3,4,5,6,7,8],dtype=np.float)
T2,x2,y2,px2,py2,E2,TT2,RAD2,Xi2 = np.loadtxt(rf"{run_dir}/Data2/orbt1q"
                                              +str(file2).zfill(3)+".dat",unpack=True,
                                              usecols=[0,1,2,3,4,5,6,7,8],dtype=np.float)
```

```python
file1=1
file2=2
fig, ax1 = plt.subplots()

#Time evolution of electron energy
T1,x1,y1,px1,py1,E1,TT1,RAD1,Xi1 = np.loadtxt(rf"{run_dir}/Data1/orbt1q"
                                              +str(file1).zfill(3)+".dat",unpack=True,
                                              usecols=[0,1,2,3,4,5,6,7,8],dtype=np.float)
T2,x2,y2,px2,py2,E2,TT2,RAD2,Xi2 = np.loadtxt(rf"{run_dir}/Data2/orbt1q"
                                              +str(file2).zfill(3)+".dat",unpack=True,
                                              usecols=[0,1,2,3,4,5,6,7,8],dtype=np.float)
T1 = T1*1e15 
T2 = T2*1e15 
E1 = E1*1e-6
E2 = E2*1e-6
ax1.plot(T1,E1,color='blue', label='Classical')
ax1.plot(T2,E2,color='red', label='QED-assisted')

ax1.set_ylabel("Energy (MeV)")
ax1.set_xlabel("t (fs)")

ax1.minorticks_on()

ax1.legend()
fig = plt.gcf()
fig.set_size_inches(fig_width*2,fig_width*1.618)
fig.tight_layout()
```

# Radiation angular distribution

```python
file1 = '52'
file2 = '53'
X = np.arange(128)
Y = np.arange(128)
gs = gridspec.GridSpec(1,2)
#*****************************************************
ax= plt.subplot(gs[0,0])
#*****************************************************
angy,angz,dist_fn = np.loadtxt(rf"{run_dir}/Data{file1}/phtnTe0{file1}.dat",unpack=True,usecols=[0,1,2],dtype=np.float)
X = angz*1000
Y = angy*1000 + 440
z = dist_fn

N = int(len(z)**.5)
z = z.reshape(N, N)
extent = np.min(X), np.max(X), np.min(Y), np.max(Y)

img = ax.imshow(z,extent=extent,interpolation='nearest',
            aspect='auto',cmap='jet',
            origin='lower')
cbar=plt.colorbar(img,ticklocation='right')
#cbar.set_label(r'$dE/d\theta_yd\theta_z~\mathrm{[a.u]}$')
cbar.ax.tick_params(labelsize=7)
cbar.ax.minorticks_on()
plt.text(-90,350,r'(a)')
plt.text(-90,-450,r'Classical',fontsize=7)    
ax.set_ylabel(r'$\theta_y~\mathrm{[mrad]}$')
ax.set_xlabel(r'$\theta_z~\mathrm{[mrad]}$')
ax.set_ylim(-500,500)
ax.set_xlim(-100,100)
ax.set_xticks([-80,0,80])
ax.minorticks_on()

#*****************************************************
ax= plt.subplot(gs[0,1])
#*****************************************************
angy,angz,dist_fn = np.loadtxt(rf"{run_dir}/Data{file2}/phtnTe0{file2}.dat",unpack=True,usecols=[0,1,2],dtype=np.float)
X = angz*1000
Y = angy*1000 + 440
z = dist_fn

N = int(len(z)**.5)
z = z.reshape(N, N)
extent = np.min(X), np.max(X), np.min(Y), np.max(Y)

img = ax.imshow(z,extent=extent,interpolation='nearest',
            aspect='auto',cmap='jet',
            origin='lower')
cbar=plt.colorbar(img,ticklocation='right')
cbar.set_label(r'$dE/d\theta_yd\theta_z~\mathrm{[a.u]}$',fontsize=6)
cbar.ax.tick_params(labelsize=7)
cbar.ax.minorticks_on()
ax.set_yticklabels([])
plt.text(-90,350,r'(b)')
plt.text(-90,-450,r'QED-assisted',fontsize=7)     
ax.set_xlabel(r'$\theta_z~\mathrm{[mrad]}$')
ax.set_ylim(-500,500)
ax.set_xlim(-100,100)
ax.set_xticks([-80,0,80])
ax.minorticks_on()
fig = plt.gcf()
fig.set_size_inches(fig_width,1.4)
fig.tight_layout()
plt.show()
```

```python
fig = plt.figure()
ax1 = plt.axes()
#ax1.grid(which='major', axis='both')
file1=52
file2=53
file3=54
file4=55
ENE,NUM1,NUM2 = np.loadtxt(rf"{run_dir}/Data{file1}/phtne"+str(file1).zfill(3)+".dat",unpack=True,usecols=[0,1,2],dtype=np.float)
ENE = ENE*1e-6
ax1.plot(ENE,NUM1*10,color='blue', label='Classical, $\sigma=5~\mathrm{\mu m}$')
ENE,NUM1,NUM2 = np.loadtxt(rf"{run_dir}/Data{file2}/phtne"+str(file2).zfill(3)+".dat",unpack=True,usecols=[0,1,2],dtype=np.float)
ENE = ENE*1e-6
ax1.plot(ENE,NUM1*10,color='red',label=r'QED-assisted, $\sigma=5~\mathrm{\mu m}$')
ENE,NUM1,NUM2 = np.loadtxt(rf"{run_dir}/Data{file3}/phtne"+str(file3).zfill(3)+".dat",unpack=True,usecols=[0,1,2],dtype=np.float)
ENE = ENE*1e-6
ax1.plot(ENE,NUM1*10,color='blue',label=r'Classical, $\sigma=15~\mathrm{\mu m}$',linestyle='--')
ENE,NUM1,NUM2 = np.loadtxt(rf"{run_dir}/Data{file4}/phtne"+str(file4).zfill(3)+".dat",unpack=True,usecols=[0,1,2],dtype=np.float)
ENE = ENE*1e-6
ax1.plot(ENE,NUM1*10,color='red',label='QED-assisted, $\sigma=15~\mathrm{\mu m}$',linestyle='--')
ax1.set_xlabel("Energy [MeV]")
ax1.set_ylabel(r"$\mathrm{(dN/dE) \times dE}~\mathrm{(10~MeV~resolution)}$")
ax1.set_yscale('log')
ax1.set_xlim(xmin=0,xmax=600)
ax1.set_ylim(ymin=1e3,ymax=1e10)
ax1.minorticks_on()
ax1.legend(loc=1)
fig = plt.gcf()
fig.set_size_inches(fig_width,2.6)
fig.tight_layout()
plt.show()
```
